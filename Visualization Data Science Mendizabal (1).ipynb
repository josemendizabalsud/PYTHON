```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
Base = pd.read_csv("C:/jose/HRDataset_v14.csv")
Base
```

```python
#Estadística descriptiva
Base.describe()
```

```python
#Histograma salario
fig, ax = plt.subplots(figsize=(10, 4))
ax.hist(Base["Salary"].values.flatten(), bins=5,ec="black",color="grey")
ax.set_title('Histograma de precipitaciones')
ax.set_xlabel('Intervalos de precipitaciones (mm.)')
ax.set_ylabel('Frecuencia absoluta')

plt.show()
```

```python
#Gráfico violin salario discriminado género
ax = sns.catplot(data=Base, kind='violin', x='Sex', y='Salary', hsplit=False)
ax.set(xlabel='Género', ylabel='Salario', title='Salario por genero')

plt.show()
```

```python
#
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(Base['DateofHire'],Base.index, label='Ingreso personal')




ax.set_xlabel('Año')  
ax.set_ylabel('Personal') 
ax.set_title('Personal por año') 	
ax.set_xlim(Base.index[0], Base.index[-1])
ax.legend() 	

plt.show()
```
