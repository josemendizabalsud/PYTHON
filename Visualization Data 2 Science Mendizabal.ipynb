```python
df_ejercicio = sns.load_dataset('exercise')
df_ejercicio = df_ejercicio.drop('Unnamed: 0', axis='columns')
df_ejercicio.head()
```

```python
df_30_min = df_ejercicio[df_ejercicio['time'] == '30 min']
df_30_min.head()
```

```python
plt.figure()
ax = sns.displot(data=df_30_min, x='pulse', kind='kde', hue='kind', fill=True)
ax.set(xlabel='Frecuencia Cardaca', ylabel='Densidad', title='Distribucin  de las pulsaciones')
plt.show()
```

```python
plt.figure()
ax = sns.displot(data=df_30_min, x='pulse', kind='kde', hue='diet', fill=True)
ax.set(xlabel='Frecuencia Cardaca', ylabel='Densidad', title='Distribucin  de las pulsaciones')
plt.show()
```

```python
ax = sns.catplot(data=df_ejercicio, kind='violin', x='time', y='pulse', hue='diet', split=False)
ax.set(xlabel='Duracin de ejercicio', ylabel='Frecuencia cardaca', title='Categorizacin de la distribucin de pulsaciones')

plt.show()
```
